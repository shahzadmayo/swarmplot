# Python â€“ seaborn swarmplot
### Draw a categorical scatterplot with non-overlapping points.

```python
# import library
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = sns.load_dataset("tips")

# plot the swarmplot
# size set to 5
sns.swarmplot(x ="day", y = "total_bill",
			data = data, size = 5)
plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("tips")

# plot the swarmplot
# hue by size
# oriented to horizontal
sns.swarmplot(y = "day", x = "total_bill", hue = "size",
			 orient = "h",data = data)
plt.show()
```

```python


sns.set(style='whitegrid')
fmri = sns.load_dataset("fmri")

sns.swarmplot(x="timepoint",
				y="signal",
				data=fmri)
```

```python
sns.set(style='whitegrid')
fmri = sns.load_dataset("fmri")

sns.swarmplot(x="timepoint",
				y="signal",
				hue="region",
				data=fmri)
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.swarmplot(x=tips["total_bill"])
```

```python
sns.swarmplot(y="total_bill", x="day", data=tips,
				linewidth=2,edgecolor='green')
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.swarmplot(x="day", y="total_bill", hue="smoker",
				data=tips, palette="Set2", dodge=True)
```

# Plotting large points and different aesthetics With marker and alpha parameter:

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.swarmplot(x="day", y="total_bill", hue="smoker",
				data=tips, palette="Set2", size=20, marker="D",
				edgecolor="gray", alpha=.25)
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.swarmplot(x="time", y="tip", data=tips,
				order=["Dinner", "Lunch"])
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.swarmplot(x='day', y='total_bill', data=tips,
				hue='smoker', size=10)
```
